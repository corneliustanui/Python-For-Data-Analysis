```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(-10, 10, .2)
Y = np.linspace(0, len(X), len(X))
```

```python
plt.plot(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('input data')
plt.show()
```

```python
def sigmoid(arr, offset):
    a = []
    for x in arr:
        a.append(1 / (1+np.exp(-x+offset)))
    return a

S = sigmoid(X, 0)
```

```python
plt.plot(X, S)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('sigmoid')
plt.show()
```

```python
nb_sum = 10 # number of logistics to sum
offsets = np.linspace(0, np.max(X), 100)

S_sum = np.zeros(np.shape(X))

for offset in offsets:
    S_sum += sigmoid(X, offset)

S_sum = S_sum / nb_sum
```

```python
plt.plot(X, S_sum)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sum of sigmoids')
plt.show()
```

```python
def sigmoid_approx(x):
    return np.log(1 + np.exp(x))

S_approx = sigmoid_approx(X)
```

```python
plt.plot(X, S_approx)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('approximation of the sum of sigmoids')
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

relu1 = relu(X)
```

```python
plt.plot(X, relu1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ReLU')
plt.show()
```

```python
def elu(arr, alpha):
    a = []
    for x in arr:
        if x >= 0:
            a.append(x)
        else:
            a.append(alpha * (np.exp(x)-1))
    return a

elu1 = elu(X, 1.0)
```

```python
plt.plot(X, elu1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ELU')
plt.show()
```
